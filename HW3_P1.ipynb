###usage of 'random_state = 42':

The random_state parameter in the train_test_split function in Scikit-learn is used to set the seed for the random number generator, ensuring that the train and test datasets are split in the same way every time the code is executed. The value of random_state is arbitrary, and 42 is commonly used as a convention, based on its significance in popular culture as the "Answer to the Ultimate Question of Life, the Universe, and Everything" in Douglas Adams' science fiction novel "The Hitchhiker's Guide to the Galaxy."
Using a fixed value for random_state ensures that the results of the model are reproducible, allowing for more consistent and reliable comparisons between different models and experiments. It is recommended to set the random_state parameter in Scikit-learn algorithms, including train_test_split, to ensure that the results are reproducible and can be validated by others.

You can set the random_state parameter to any integer value. The specific value you choose for random_state does not affect the quality of the split; it only determines the random seed used for the splitting process.

The difference between setting the random seed to 42 or 12 or any other number is that the sequence of random numbers generated will be different, but the property of reproducibility will remain the same.

```python
from tensorflow import keras
from tensorflow.keras import initializers

def build_model(input_shape, hidden_size, output_size, number_of_hidden_layers, activation_func, initializer):
  # Define model with sequential API
  model = keras.Sequential()

  # Input layer
  model.add(keras.Input(shape=input_shape))

  # Hidden layers with tanh activation (Xavier initialization)
  for _ in range(number_of_hidden_layers):
    model.add(keras.layers.Dense(hidden_size, activation=activation_func, kernel_initializer=initializer))

  # Output layer with softmax activation
  model.add(keras.layers.Dense(output_size, activation="softmax"))

  # Compile the model (optimizer, loss function, metrics)
  model.compile(optimizer="sgd", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
  return model

# Example usage
# model = build_model((784,), 1000, 10)  # Assuming input size 784 (e.g., 28x28 images), 1000 hidden units, 10 output classes
# ... (train the model using your data and chosen learning algorithm)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def display_loss(history):
  plt.plot(history.history['loss'])
  plt.plot(history.history['val_loss'])
  plt.title('Model Loss')
  plt.ylabel('Loss')
  plt.xlabel('Number of Epochs')
  plt.legend(['Train', 'Test'], loc='upper left')
  plt.show()

def display_accuracy(history):
  plt.plot(history.history['accuracy'])
  plt.plot(history.history['val_accuracy'])
  plt.title('model accuracy')
  plt.ylabel('accuracy')
  plt.xlabel('no of epochs')
  plt.legend(['train', 'test'], loc='upper left')
  plt.show()
```

```python
from tensorflow import random
from tensorflow.keras import initializers

class RandomUniformInitializer(initializers.Initializer):
  def __init__(self, seed=None):
    self.seed = seed

  def __call__(self, shape, dtype=None):
    if self.seed is not None:
      random.set_seed(self.seed)
    limit = 1.0 / np.sqrt(shape[-1])
    return random.uniform(shape, minval=-limit, maxval=limit, dtype=dtype)
```

```python
#Based on the paper, page 250-251, part 2.3, we computed negative log_liklihood over 10 consecutive training pairs (x, y) and used to update parameters  in that direction
batch_size = 128
epochs = 10
```

```python
def make_relu(alpha):  # Define a custom Leaky ReLU function with learnable alpha
  def activate(x):
    return keras.activations.relu(x, alpha=alpha)
  return activate

leaky_relu = make_relu(alpha=0.01)
prelu = make_relu(alpha=0.25)
```
